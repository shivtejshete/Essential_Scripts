```python
import pandas as pd
from io import StringIO
import requests as req
import numpy as np
#!pip install fsspec
```

```python
def get_distance(Source, Destination):
  """
  Source (str) : Starting point 
  Destination (Str) : Final location

  API_KEY : Customer key generated from MAPQUEST portal
  """
  #Source= "\"MUMBAI\""
  #Destination= "\"PUNE\""
  
  Source = "\"{}\"".format(Source)
  Destination = "\"{}\"".format(Destination)

  Key_string = "{"+"\"locations\""+":"+"["+"{}".format(Source)+","+"{}".format(Destination)+"]"+"}"
  Key_string= 'http://open.mapquestapi.com/directions/v2/optimizedroute?key=API_KEY&json={}'.format(Key_string)

  resp = req.get(Key_string)
  newdf = pd.read_json(StringIO(resp.text))
  distance_miles = newdf['route'][3]
  distance_km = distance_miles * 1.60934
  print("Distance between {} -and- {} = {} km ".format(Source, Destination, distance_km))
  return distance_km
```
